You can read in [tabular data really easily](http://pandas.pydata.org/pandas-docs/dev/io.html) from a csv file, but let me just generate a random 100 x 100 matrix of floats (eg a numpy array).

```python
# Import several useful plotting and numerical utilitiles
%pylab inline 


# Generate 100x100 matrix using numpy (np)
data = np.random.randn(100,100)
print 'Shape of data is', data.shape
```

Read the numpy matrix into a `pandas.DataFrame`.  We can add some row and column labels as well.  In fact, let me just generate random labels (http://stackoverflow.com/questions/2257441/random-string-generation-with-upper-case-letters-and-digits-in-python).

```python
import string
import random

def id_generator(size=6, chars=string.ascii_uppercase + string.digits):
    return ''.join(random.choice(chars) for _ in range(size))

row_labels = [id_generator() for i in range(100)]
col_labels = [id_generator() for i in range(100)]
```

```python
import pandas as pd
import string

df = pd.DataFrame(data, index=row_labels, columns=col_labels)
df
```

We can slice and plot really easily.  For example. plotting the first five columns:

```python
df.iloc[::, 0:5].plot()
plt.ylabel('value');
```

Correlation/covariance and most other statistical functions are builtin

```python
correlations = df.corr()
correlations
```

We can use `imshow`, which is a plotting function for a 2d matrix (actually it's for images), to view the 2d map of this matrix.

```python
imshow(correlations, cmap='hot')
plt.title('Correlations in random data')
plt.colorbar();
```
